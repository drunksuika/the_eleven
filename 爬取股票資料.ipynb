```python
import requests as r #requests 向台灣證券交易所 API 發送請求
import json
import pandas as pd
from datetime import datetime, date
import matplotlib.pyplot as plt #用於將股價資訊視覺化
datestr = input("輸入要查詢股票資訊的日期（例如：20220101）")
tsno = input("輸入要查詢股票代號（例如：2330）")
%matplotlib inline
url = "https://www.twse.com.tw/exchangeReport/STOCK_DAY?response=json&date=" + datestr + "&stockNo=" + tsno
r.get(url)
res = r.get(url)
res.json()
stock_json = res.json()
stock_json['data']
pd.DataFrame.from_dict(stock_json['data']) #將 JSON 格式的資料轉換成易於處理的 DataFrame
stock_df = pd.DataFrame.from_dict(stock_json['data'])
stock_df.head()
stock_json['fields']
stock_df.columns = stock_json['fields']
stock_df.head()
```
